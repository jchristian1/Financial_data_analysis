```python
import pandas as pd
```

```python
pd.set_option('display.max_rows',None,'display.max_columns',None)
```

```python
eps = pd.read_csv('50-eps.csv',index_col = 0)
```

```python
eps
```

```python
## let's check variables type
eps.transpose()
```

```python
eps = eps.transpose() 
```

```python
eps.head()
```

```python
eps['10Y CAGR'] = (eps[2019]/eps[2009])**(1/10) - 1
```

```python
eps['5Y CAGR'] = (eps[2019]/eps[2014])**(1/5) - 1
```

```python
eps['1Y CAGR'] = (eps[2019]/eps[2018])**(1) - 1
eps
```

```python
eps.sort_values('10Y CAGR', ascending=False)
```

```python
data = eps[['10Y CAGR','5Y CAGR','1Y CAGR']]
```

```python
data.head()
```

```python
data.to_csv('result-eps.csv')
```
