```python
import pandas as pd
```

```python
#let's load our debth to equity dataset
de = pd.read_csv('50-de-ratio.csv',index_col=0)
```

```python
#Let's try to see the entire dataset
pd.set_option('display.max_rows',None,"display.max_columns",None)
```

```python
## Lets see our full debth to equity dataset
de.index
```

```python
##we can see our index is data type object what means we need to parse it into a date_time type
de.index = pd.to_datetime(de.index)
```

```python
## Now we have our dates in a correct format
de.index
```

```python
## Let's sort all the dataset by index
de = de.sort_index(ascending=False)
de
```

```python
## Let's get just the last 10 years of the dataset to do our analysis
de = de.loc['2010-01-01':]
de
```

```python
##let's create a time series quaterly using the mean
de.resample('Q', convention='end').agg('mean').head()
```

```python
# Let's transpose the matrix so it will be easier to read
de.resample('Q', convention='end').agg('mean').transpose()
```

```python
## Let's create a list of debth to equity ratio at the laste year with more information
col = de.resample('Q', convention='end').agg('mean').transpose()['2020-09-30']
col
```

```python
## we are gonna be interested just in those companies that have a D/E less than 5 (for investing purposes, they are less risky, and bigger than 0)
result = (col < 5) & (col > 0)
result
```

```python
result.name = "D/E ratio"
result.to_csv("result-de-.csv")
```
