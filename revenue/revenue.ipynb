```python
rev = pd.read_csv('50-rev.csv',index_col=0)
```

```python
rev
```

```python
## year 2020 have a lot of missing values so we will work with just 2009 to 2019
rev = rev.loc['2009':'2019']
rev
```

```python
rev = rev.transpose()
rev
```

```python
## let's calculate the 10 anual growth rate
## compound annual growth 10 years
rev['10Y CAGR'] = ((rev[2019]/rev[2009])**(1/10) - 1)
```

```python
## let's calculate the 10 anual growth rate
## compound annual growth 10 years
rev['5Y CAGR'] = ((rev[2019]/rev[2014])**(1/5) - 1)
```

```python
## let's calculate the 10 anual growth rate
## compound annual growth 10 years
rev['1Y CAGR'] = ((rev[2019]/rev[2018])**(1) - 1)
```

```python
rev.head()
```

```python
## Let's create a new matrix with just the calculated columns
data = rev[['10Y CAGR','5Y CAGR','1Y CAGR']]
```

```python
data.head()
```

```python
##let's sort the data by 10 years compound annual growth
data.sort_values('10Y CAGR', ascending=False)
```

```python
data.to_csv('result-rev.csv')
```
