```python
import pandas as pd
```

```python
pd.set_option('display.max_rows',None,'display.max_columns',None)
```

```python
bvps = pd.read_csv('50-bvps.csv',index_col=0)
bvps
```

```python
##let's format the index
bvps.index = pd.to_datetime(bvps.index)
bvps.index
```

```python
##Let's sort the data by date
bvps.sort_index(ascending=False,inplace=True)
```

```python
bvps.head()
```

```python
## quaterly mean
bvps = bvps.resample('Q', convention='end').agg('mean')
```

```python
## Let's create the new dataset for las 10 years
bvps = bvps.loc['2010-09-01':]
```

```python
bvps =bvps.transpose()
bvps
```

```python
##lets create the dataset we are going to export later with the new calculated fields
result = pd.DataFrame(data=None,index=bvps.index)
```

```python
result['10Y CAGR'] = (bvps['2020-09-30']/bvps['2010-09-30'])**(1/10) -1
```

```python
result['5Y CAGR'] = (bvps['2020-09-30']/bvps['2015-09-30'])**(1/5) -1
```

```python
result['1Y CAGR'] = (bvps['2020-09-30']/bvps['2019-09-30'])**(1) -1
```

```python
## let's see the growth of each company for the last 10 years sorted
result.sort_values('10Y CAGR',ascending=False)
```

```python
## let's see the growth of each company for the last 5 years sorted
result.sort_values('5Y CAGR',ascending=False)
```

```python
## let's see the growth of each company for the last year sorted
result.sort_values('1Y CAGR',ascending=False)
```

```python
## export New dataset
result.to_csv('result-bvps.csv')
```
